```python
#the top performimg products in the usa on the amazon plartform 
filtered = df[(df['Location'] == 'USA') & (df['Platform'] == 'Amazon')]

product_sales = filtered.groupby('Product Name')['Sales'].sum()

top_product = product_sales.idxmax()
top_value = product_sales.max()

print(f"Top product: {top_product} with sales of {top_value}")
```
