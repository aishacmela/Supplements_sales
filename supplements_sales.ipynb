```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dt = pd.read_csv('supplements_sales_data.csv', sep=';')
dt.head()
```

```python
dt.describe()
```

```python
print(dt.dtypes)
```

```python
dt['Date'] = pd.to_datetime(dt['Date'])
```

```python
#strip whitespace fic capitalisation
for col in ['Product Name', 'Category', 'Location', 'Platform']:
    dt[col] = dt[col].str.strip().str.title()


#checking missing values 
dt_missing = dt.isnull().sum()
print(dt_missing)

#duplicates 
dubes = dt.duplicated().sum()
print(dubes)

```

```python
#get a propotion of products that came back 
dt['Return rates'] = dt['Units Returned'] / dt['Units Sold']

#calculates what the customer actually paid after discount
dt['Effective Price'] = dt['Price'] * (1 - dt['Discount'])

#divides total revenue by units sold to the average earning per item sold
dt['Revenue Per Unit'] = dt['Revenue'] / dt['Units Sold']
print(dt.head(10))
```
